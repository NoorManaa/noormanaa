```python
import numpy as np # linear algebra import pandas as pd
import pandas as pd
from sklearn.ensemble import RandomForestRegressor 
from sklearn.model_selection import train_test_split

traindata=pd.read_csv('/kaggle/input/decision-tree-sample/train.csv') 
testdata=pd.read_csv('/kaggle/input/decision-tree-sample/test.csv')

traindata.drop('id',axis=1,inplace=True)

id=testdata['id'] 
testdata.drop('id',axis=1,inplace=True)

x=traindata.drop("Response",axis=1)
y=traindata["Response"]

x_train, _, y_train, _= train_test_split(x, y, test_size=0.2, random_state=42) 
rf_model = RandomForestRegressor(n_estimators=180, random_state=66)

rf_model.fit(x_train, y_train)

y_predict = rf_model.predict(testdata)

predict_data=pd.DataFrame({'id':id,'Response':y_predict})

predict_data.to_csv('randomforest9.csv',index=False)
```
